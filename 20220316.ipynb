#sine curve 생성

```python
import matplotlib.pyplot as plt

plt.figure()

theta1 = np.linspace(0, 2*np.pi, 100)
s1 = np.sin(theta1)
theta2 = np.linspace(0, 2*np.pi, 100) + np.pi/2
s2 = np.sin(theta2)

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(theta1, s1, '.')
plt.subplot(2, 1, 2) # (rows, columns, panel number)
plt.plot(theta2, s2, '.')
```

theta1 = np.linspace(0, 2*np.pi, 100) : 0부터 2pi까지를 100등분한 후 theta1이라는 변수에 assign
s1 = np.sin(theta1) : s1이라는 변수에 theta1을 싸인함수화

theta2 = np.linspace(0, 2*np.pi, 100) + np.pi/2 : theta1값 +pi/2값
s2 = np.sin(theta2) : theta2에 해당하는 싸인함수

subplot은 그래프를 시각화할 때 보여줄 그래프 화면의 개수
plt.subplot(2,1,1) : 화면을 2*1로 쪼갠 뒤 1번째 그래프를 보여주는 함수

plot.plot(theta1,s1,'.') : theta1은 x값, s1은 y값으로 그래프 생성

```python
np.corrcoef(s1,s2)
```

corrcoef : correlation coefficient. 함수 s1과 s2 사이의 상관계수

결과값에서 첫 값과 마지막 값은 보지말고

중간값 2개를 보면 같은 수. 2.410xx * 10^-17 = 사실상 0

즉 상관계수가 거의 0이므로 s1과 s2는 전혀 관련이 없음. 상관계수가 -1이면 역방향, 1이면 정방향, 0 이면 무관한 관계
